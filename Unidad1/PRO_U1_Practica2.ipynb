**Practica 2: Maximo, minimo, rango y dispersion de datos**

```python
import random
import matplotlib.pyplot as plt
from statistics import *
```

```python
fabrica1 = []
for i in range(100):
  numero_aleatorio = round (random.uniform(50, 100),2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2 = []
for i in range(100):
  numero_aleatorio = round (random.uniform(40, 93),2)
  fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
maximo1 = max(fabrica1)
minimo1 = min(fabrica1)
rango1 = maximo1 - minimo1
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)

print("Estadistica descriptiva de la fabrica 1 \n", "Media =", media1,
      "\n Mediana = ", mediana1, "\n Moda =", moda1, "\n Maximo = ", maximo1, "\n Minimo =", minimo1, "\n Rango =", rango1)
```

```python
maximo2 = max(fabrica2)
minimo2 = min(fabrica2)
rango2 = maximo2 - minimo2
media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)

print("Estadistica descriptiva de la fabrica 2 \n", "Media =", media2,
      "\n Mediana = ", mediana2, "\n Moda =", moda2, "\n Maximo = ", maximo2, "\n Minimo =", minimo2, "\n Rango =", rango2)
```

```python
desviancion_estandar_Fabrica1 = stdev(fabrica1)
print(desviancion_estandar_Fabrica1)
```

```python
desviancion_estandar_Fabrica2 = stdev(fabrica2)
print(desviancion_estandar_Fabrica2)
```

```python
plt.figure(figsize=(8,6))
plt.hist(fabrica1, bins=10 , alpha=0.2, color='blue', edgecolor = 'black')
plt.axvline(media1, color='blue', linestyle='dashed', linewidth=2, label=f'media={media1:.2f}')
plt.axvline(mediana1, color='red', linestyle='dashed', linewidth=2, label=f'mediana={mediana1:.2f}')
plt.axvline(moda1, color='yellow', linestyle='dashed', linewidth=2, label=f'moda={moda1:.2f}')
plt.axvline(media1 - desviancion_estandar_Fabrica1, color='black', linestyle='dashed', linewidth=2, label=f'desviacion estandar={desviancion_estandar_Fabrica1:.2f}')
plt.axvline(media1 + desviancion_estandar_Fabrica1, color='black', linestyle='dashed', linewidth=2)
plt.xlabel('valor')
plt.ylabel('frecuencia')
plt.title('produccion diaria de la fabrica 1')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(fabrica2, bins=10 , alpha=0.2, color='blue', edgecolor = 'black')
plt.axvline(media2, color='blue', linestyle='dashed', linewidth=2, label=f'media={media2:.2f}')
plt.axvline(mediana2, color='red', linestyle='dashed', linewidth=2, label=f'mediana={mediana2:.2f}')
plt.axvline(moda2, color='yellow', linestyle='dashed', linewidth=2, label=f'moda={moda2:.2f}')
plt.axvline(media2 - desviancion_estandar_Fabrica2, color='black', linestyle='dashed', linewidth=2, label=f'desviacion estandar={desviancion_estandar_Fabrica2:.2f}')
plt.axvline(media2 + desviancion_estandar_Fabrica2, color='black', linestyle='dashed', linewidth=2)
plt.xlabel('valor')
plt.ylabel('frecuencia')
plt.title('produccion diaria de la fabrica 2')
plt.legend()
plt.show()
```
