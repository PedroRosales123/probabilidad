```python
import math as mt
from itertools import permutations
asistentes = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"]
n = len(asistentes)
r = 4
Pr = mt.factorial(n) / mt.factorial(n - r)
print("Existen", int(Pr), "formas diferentes de asignar asistentes a las preguntas.\n")
print("Aquí están TODAS las permutaciones posibles:\n")
permutaciones_asistentes = list(permutations(asistentes, r))
for i, perm in enumerate(permutaciones_asistentes, start=1):
    print(f"{i}: {perm}")
```
