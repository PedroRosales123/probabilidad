```python
# ===============================================================
# PROYECTO FINAL: Análisis Estadístico de Datos COVID-19
# ===============================================================


# ------------------ IMPORTACIÓN DE LIBRERÍAS ------------------
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
sns.set(style="whitegrid")


# ---------------------- RUTA DEL ARCHIVO -----------------------
ruta = "/content/drive/MyDrive/datos_covid/covid2020-2021-codificado.csv"

# ---------------------- CARGA DEL ARCHIVO ----------------------
# Cargar sin dtype para evitar errores por valores no numéricos
columnas = [
    "SECTOR","SEXO","ENTIDAD_NAC","TIPO_PACIENTE","FECHA_INGRESO","FECHA_DEF",
    "INTUBADO","NEUMONIA","EDAD","EMBARAZO","INDIGENA","DIABETES","EPOC","ASMA",
    "HIPERTENSION","CARDIOVASCULAR","OBESIDAD","RENAL_CRONICA","TABAQUISMO",
    "RESULTADO_ANTIGENO","UCI"
]

df = pd.read_csv(
    ruta,
    usecols=columnas,
    low_memory=True
)

# Convertir fechas
df["FECHA_INGRESO"] = pd.to_datetime(df["FECHA_INGRESO"], errors="coerce")
df["FECHA_DEF"] = pd.to_datetime(df["FECHA_DEF"], errors="coerce")

# ---------------------- VERIFICACIÓN ---------------------------
print("Dimensiones del dataset:", df.shape)
print("\nPrimeras filas:")
display(df.head())
print("\nInformación general del dataset:")
df.info()
```

```python
# ===============================================================
# PROYECTO FINAL: Análisis Estadístico de Datos COVID-19
# Segunda Parte: Estadística Descriptiva + Visualizaciones
# ===============================================================

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# ---------------------- ESTADÍSTICA DESCRIPTIVA ----------------------

# 1. Casos por resultado de antígeno (0=Negativo, 1=Positivo, 2=Pendiente)
casos_resultado = df["RESULTADO_ANTIGENO"].value_counts(dropna=True)
print("Número de casos por resultado de antígeno:\n", casos_resultado, "\n")

# 2. Pacientes hospitalizados (2) y ambulatorios (1)
pacientes_tp = df["TIPO_PACIENTE"].value_counts(dropna=True)
print("Pacientes hospitalizados y ambulatorios:\n", pacientes_tp, "\n")

# 3. Pacientes embarazadas (1 = Sí)
embarazadas = df["EMBARAZO"].value_counts().get(1, 0)
print("Pacientes embarazadas:", embarazadas, "\n")

# 4. Pacientes indígenas (1 = Sí)
indigenas = df["INDIGENA"].value_counts().get(1, 0)
print("Pacientes indígenas:", indigenas, "\n")


# --------------------------- VISUALIZACIONES ----------------------------

# A) Gráfico de barras horizontal por estado de nacimiento (ENTIDAD_NAC)
plt.figure(figsize=(10, 8))
valores = df["ENTIDAD_NAC"].value_counts().sort_values()
porcentajes = (valores / valores.sum()) * 100

bars = plt.barh(valores.index, valores, color="steelblue")

for i, v in enumerate(valores):
    plt.text(v + valores.max()*0.01, i, f"{v:,} ({porcentajes.iloc[i]:.1f}%)", va='center')

plt.title("Casos por Entidad de Nacimiento")
plt.xlabel("Número de Casos")
plt.ylabel("Entidad")
plt.show()


# HISTOGRAMA DE EDAD con número y porcentaje encima de cada barra
import numpy as np
import matplotlib.ticker as ticker

# 1) limpieza mínima de EDAD (no modifica otros datos)
edades = pd.to_numeric(df["EDAD"], errors="coerce").dropna()
edades = edades[(edades >= 0) & (edades <= 110)]  # rango humano razonable

# 2) parámetros del histograma
bins = 25
counts, bin_edges = np.histogram(edades, bins=bins)
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
total = counts.sum()

# 3) dibujar histograma (barras)
fig, ax = plt.subplots(figsize=(12,7))
bars = ax.bar(bin_centers, counts, width=(bin_edges[1]-bin_edges[0])*0.95,
              color="purple", alpha=0.8, edgecolor="black")

# 4) formatear eje Y para mostrar números completos
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

# 5) añadir etiquetas con número y porcentaje encima de cada barra
for center, count in zip(bin_centers, counts):
    if count > 0:
        pct = (count / total) * 100
        ax.text(center, count + total*0.005, f"{count:,}\n({pct:.2f}%)",
                ha='center', va='bottom', fontsize=9)

# 6) etiquetas y estilo
ax.set_title("Histograma de Edad (con número y porcentaje)")
ax.set_xlabel("Edad")
ax.set_ylabel("Frecuencia (número de pacientes)")
ax.grid(axis="y", alpha=0.3)
plt.show()


# C) Línea: Evolución de Casos por Fecha de Ingreso
serie_fechas = df.groupby("FECHA_INGRESO").size()

plt.figure(figsize=(14, 6))
plt.plot(serie_fechas.index, serie_fechas.values)
plt.title("Evolución de Casos por Fecha de Ingreso")
plt.xlabel("Fecha de Ingreso")
plt.ylabel("Número de Casos")
plt.grid(True)
plt.show()


# D) Gráfico de barras apiladas por sexo y resultado
tabla = pd.crosstab(df["SEXO"], df["RESULTADO_ANTIGENO"])

ax = tabla.plot(kind="bar",
                stacked=True,
                figsize=(10, 6),
                colormap="viridis")

# --- Etiquetas arriba de cada barra ---
for pos, (index, row) in enumerate(tabla.iterrows()):
    total = row.sum()
    ax.text(pos, total + tabla.values.max()*0.01,
            f"{total:,}",
            ha="center", fontsize=12)

plt.title("Resultado de Prueba por Sexo (Apilado)")
plt.xlabel("Sexo")
plt.ylabel("Número de Casos")
plt.legend(title="Resultado")
plt.xticks(rotation=0)
plt.show()


# E) Gráfico circular por tipo de paciente
tp_counts = df["TIPO_PACIENTE"].value_counts(dropna=True)

plt.figure(figsize=(8, 8))

# Crear etiquetas con número + porcentaje
labels = [
    f"{categoria} ({valor} casos, {valor/tp_counts.sum()*100:.1f}%)"
    for categoria, valor in zip(tp_counts.index, tp_counts.values)
]

plt.pie(tp_counts,
        labels=labels,
        startangle=90)

plt.title("Distribución por Tipo de Paciente")
plt.show()


# F) Gráfico de barras por comorbilidad
comorbilidades = ["DIABETES","EPOC","ASMA","HIPERTENSION",
                  "CARDIOVASCULAR","OBESIDAD","RENAL_CRONICA","TABAQUISMO"]

conteo_comorb = {col: df[col].value_counts().get(1, 0) for col in comorbilidades}

serie_comorb = pd.Series(conteo_comorb).sort_values()
total_comorb = serie_comorb.sum()
porcentajes = (serie_comorb / df.shape[0]) * 100

plt.figure(figsize=(12, 6))
bars = plt.bar(serie_comorb.index, serie_comorb.values, color="darkred")

plt.title("Pacientes con Comorbilidades")
plt.ylabel("Número de Casos")

# Etiquetas arriba de cada barra
for i, (valor, pct) in enumerate(zip(serie_comorb.values, porcentajes)):
    plt.text(i, valor + serie_comorb.max()*0.01,
             f"{valor:,}\n({pct:.2f}%)",
             ha='center')

plt.xticks(rotation=60)
plt.show()
```

```python
# ===============================================================
# PROBABILIDADES POR COMORBILIDAD E INTERPRETACIÓN ESTADÍSTICA
# ===============================================================

# Lista exacta de comorbilidades
comorbilidades = [
    "DIABETES", "HIPERTENSION", "OBESIDAD", "EPOC",
    "ASMA", "CARDIOVASCULAR", "RENAL_CRONICA", "TABAQUISMO"
]

total_general = df.shape[0]
print(f"\nTotal de pacientes en la base de datos: {total_general}\n")

# --------------------------------------------------------------
# 1. PROBABILIDAD SIMPLE
# --------------------------------------------------------------

print("===== PROBABILIDAD SIMPLE DE CADA COMORBILIDAD =====\n")

prob_simple = {}

for comorb in comorbilidades:
    num = (df[comorb] == "Si").sum()     # usa "Si" / "No" exactamente como están
    prob = num / total_general
    prob_simple[comorb] = prob

    print(f"{comorb}: {num} casos → P({comorb}) = {prob:.4f} ({prob*100:.2f}%)")

# --------------------------------------------------------------
# 2. PROBABILIDAD CONDICIONAL POR ENTIDAD_NAC
#    P(comorbilidad | entidad)
# --------------------------------------------------------------

print("\n\n===== PROBABILIDAD CONDICIONAL POR ENTIDAD DE NACIMIENTO =====\n")

prob_condicional = {}

for comorb in comorbilidades:
    # Conteo de "Si" por estado
    tabla = df.groupby("ENTIDAD_NAC")[comorb].apply(lambda x: (x == "Si").sum())

    # Total de registros por estado
    total_por_entidad = df["ENTIDAD_NAC"].value_counts()

    # Probabilidad condicional
    prob_por_entidad = tabla / total_por_entidad

    prob_condicional[comorb] = prob_por_entidad

    print(f"\n--- {comorb} ---")
    print(prob_por_entidad.round(4))

# --------------------------------------------------------------
# 3. INTERPRETACIÓN ESTADÍSTICA
# --------------------------------------------------------------

print("\n\n===== INTERPRETACIÓN ESTADÍSTICA =====\n")

# ---- Interpretación Simple ----
for comorb in comorbilidades:
    p = prob_simple[comorb]

    if p < 0.01:
        nivel = "muy baja"
    elif p < 0.05:
        nivel = "baja"
    elif p < 0.15:
        nivel = "moderada"
    else:
        nivel = "alta"

    print(f"La probabilidad simple de {comorb} es {p:.3f} ({p*100:.1f}%), "
          f"lo que indica una prevalencia {nivel} entre los pacientes analizados.")

# ---- Interpretación por entidad ----
print("\nResumen por entidad:\n")

for comorb in comorbilidades:
    serie = prob_condicional[comorb].dropna()

    estado_max = serie.idxmax()
    estado_min = serie.idxmin()

    print(f"\n• {comorb}:")
    print(f"  - Mayor probabilidad: {estado_max} ({serie.max()*100:.1f}%)")
    print(f"  - Menor probabilidad: {estado_min} ({serie.min()*100:.1f}%)")
```

```python
from scipy.stats import norm

# Parámetros del problema
mu = 2360       # media
sigma = 714     # desviación estándar

# Probabilidad de que X esté entre 2000 y 3000
p = norm.cdf(3000, mu, sigma) - norm.cdf(2000, mu, sigma)

# Mostrar resultado
print("""En un análisis de casos positivos de COVID por estado, se observa que el número de
casos sigue una distribución normal con:
• Media: 2360 casos
• Desviación estándar: 714 casos
Pregunta: ¿Cuál es la probabilidad de que un estado tenga entre 2000 y 3000 casos positivos?
""")
print(f"La probabilidad es: {p:.4f} ({p*100:.2f}%)")
```
